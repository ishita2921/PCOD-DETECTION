#### Data Cleaning

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\HP\PCOD-DETECTION\data\PCOS_data.csv")
df.head()
```

```python
#check for missing values
missing_summary=df.isnull().sum()

# Strategy: Remove rows/columns with too many nulls or fill with mean/median for numeric
# fill numeric with median and categorical with mode
for column in df.columns:
    if df[column].dtype in ['float64','int64']:
        df[column]=df[column].fillna(df[column].median())
    else:
        df[column]=df[column].fillna(df[column].mode()[0])
```

```python
#convert yes/no to 0/1
yes_no_columns=df.columns[df.isin(['yes','no','Yes','No']).any()]
for col in yes_no_columns:
    df[col]=df[col].map({'Yes':1,'No':0,'yes':1,'no':0})
    
```

```python
#Drop irrelevant column or unnamed column
columns_to_drop=[col for col in df.columns if 'unnamed' in col.lower() or 'patient_id' in col.lower()]
df.drop(columns=columns_to_drop,inplace=True)
```

```python
#Save cleaned data
df.to_csv("processed_data.csv",index=False)
```
